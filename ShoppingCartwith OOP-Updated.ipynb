### Use object-oriented programming to create a shopping cart program

```python
from IPython.display import clear_output #can use this function to clean up outputs when everything is working

#class declaration
final_list = {}
class Cart:
    def __init__(self, name, carttotal, itemlist):
        self.name = name
        self.carttotal = 0.00 #will represent the running total of the cart as items are added
        self.itemlist = {} # dictionary to add item information for each item added into cart        

    def getinfo(self):
        print(f'Welcome to your cart, {self.name}. Your cart is currently empty and ${self.carttotal:.2f}')
        
    def addItem (self):
        product = input('What item would you like to add to your cart?\n').lower()
        self.product = product
        quantity = int(input('How many would you like to add to your cart?\n'))
        self.quantity = quantity
        price = float(input('How much does this item cost?\n'))
        self.price = price
        self.carttotal = self.carttotal + (quantity * price)
        self.itemlist[self.product] = self.product, self.quantity, self.price 
        clear_output()
        if self.quantity <= 1:
            print (f'{self.product.title()} has been added to your cart.')
        elif self.quantity > 1:
            print (f'{self.quantity} {self.product.title()} have been added to your cart.')
        return
    
    def viewCart (self):
        for self.product in self.itemlist:
            print (f'You currently have {self.quantity} {self.product} in your cart.\n')
        print (f'Your current cart total is ${self.carttotal:.2f}')
        return
    
    def removeItem (self):
        self.viewCart()
        self.product = input('What would you like to remove from your cart? \n').lower()
        if self.product in self.itemlist:
            self.carttotal -= (self.quantity * self.price)
            print (f'{self.product} is being removed from your cart')
            del self.itemlist[self.product]
        elif self.product not in self.itemlist:
            print(f'There is no {self.product} in your cart.')
        return
        
    def modifyCart(self):
        self.viewCart()
        self.product = input('Which item in your cart would you like to modify?\n').lower()
        if self.product in list(self.itemlist.keys()):  
            self.newQuantity = int(input(f'What quantity do you want to change {self.product} to?\n'))
            self.carttotal = self.carttotal - (self.quantity*self.price)
            list(self.itemlist.keys().index(self.product))
            self.itemlist[list(self.itemlist.keys().index(self.product))][1] = self.newQuantity
            print (f'You\'ve successfully updated the quantity of {self.product} to {self.newQuantity}')
            self.carttotal = self.carttotal + (self.newQuantity*self.price)
            print (f' Current Cart Total: {self.carttotal}')
        else:
            print (f'There is no {self.product} in your cart.')
            
        
    def checkout (self):
        for self.product in self.itemlist:
            print (f'You have {self.quantity} {self.product} in your cart \n')
        print (f'Your current cart total is ${self.carttotal:.2f}')
        if self.carttotal >= 50:
            print ('Woohoo! You scored free 2-day shipping!')
            upgradeWanted= input('Want your package even faster? For only $2.99 more you can update to 1-day shipping! Type Yes or No') 
            if upgradeWanted.lower() == 'yes':
                self.cartTotalWithTaxes = (self.carttotal + 2.99) * 1.1
                print ('Your updated total with taxes and 1-day shipping is now {self.cartTotalWithTaxes}')
                print ('Let\'s continue onto Billing Next')
            elif upgradeWanted.lower() == 'no':
                self.cartTotalWithTaxes = (self.carttotal + 1.99) * 1.1
                print ('Your updated total with taxes and 1-day shipping is now {self.cartTotalWithTaxes}')
        elif self.carttotal < 50:
            diffToFreeShip = 50 - self.carttotal
            print (f'You\'re only ${diffToFreeShip:.2f} away from Free Shipping!')
            while diffToFreeShip > 0.00:
                addMoreToCart = input ('Would you like to add items to your cart to get Free Shipping? Enter Yes or No\n').lower()
                if addMoreToCart.lower() == 'yes':
                    self.addItem()
                    self.viewCart()
                elif addMoreToCart.lower() == 'no' or diffToFreeShip <= 0:
                    break
                else:
                    print ('Please choose a valid option')
            else:
                self.cartTotalWithTaxes = (self.carttotal + 5.99) * 1.1
                print (f'OK! Your current cart total with taxes and standard 5-day shipping is ${self.cartTotalWithTaxes}')
                upgradeWanted= input('Want your package even faster? For only $2.99 more you can update to 1-day shipping! Type Yes or No\n') 
                if upgradeWanted.lower() == 'yes':
                    self.cartTotalWithTaxes = (self.carttotal + 2.99) * 1.1
                    print (f'Your updated total with taxes and 1-day shipping is now ${self.cartTotalWithTaxes}')
                else:
                    self.cartTotalWithTaxes = (self.carttotal + 1.99) * 1.1
                    print (f'Your updated total with taxes and 1-day shipping is now ${self.cartTotalWithTaxes:.2f}')
        checkoutdone = input('Would you like to go back to the main menu or on to billing?\n')
        if checkoutdone.lower() == 'billing':
            clear_output()
            return ('Ok! Onto Billing we go!')
        else:
            return
            
    def clearcart(self):
        clear_output()
        self.itemlist.clear()

user1 = input('Welcome to our site! What should we call you? \n').title()
user = Cart (user1,0.0, 'cartitems')

clear_output()
user.getinfo()

keep_going = True
while keep_going:
    userchoice = input('What would you like to do?:\n[1] Add to Cart \n[2] Remove from Cart \n[3] View Cart \n[4] Modify Cart \n[5] Checkout \n[6] Clear Cart \n[7] Quit\n')
    if userchoice.lower() == "1" or userchoice.lower()=='add to cart' or userchoice.lower() == "add":
        clear_output()
        user.addItem()
    elif userchoice.lower()== "2" or userchoice.lower() == "remove from cart" or userchoice.lower() == "remove":
        if len(user.itemlist)> 0:
            user.viewCart()
            user.removeItem()
        else: 
            print ('You do not have anything in your cart')
    elif userchoice.lower() == "3" or userchoice.lower() == "view cart" userchoice.lower() == "view":
        if len(user.itemlist) == 0:
            print ('You do not have anything in your cart right now.')
        else:    
            clear_output()
            user.viewCart()
    elif userchoice.lower() == "4" or userchoice.lower() == "modify cart" or userchoice.lower() == "modify":
        user.modifyCart()
    elif userchoice.lower() == "5" or userchoice.lower() == "checkout":
        user.checkout()
    elif userchoice.lower() == "6" or userchoice.lower() == "clear cart" userchoice.lower() == "clear":
        user.clearcart()
        print ('Your cart is now cleared out')
    elif userchoice.lower() == "7" or userchoice.lower() == "quit":
        keep_going = False
        print ('We\'ll be here when you\'re ready for your items {user1}')
    else: 
        print('Please pick a valid option')
```
